```python
import pandas as pd 
```

```python
def read_multiline_text_file(file_path):
    try:
        with open(file_path, "r") as file:
            lines = file.readlines()
            return lines
    except FileNotFoundError:
        print("File not found.")
        return None
```

```python
def convert_to_dataframe(input_txt_data, out_file_name):
    
    file_name_csv = out_file_name + '.csv'
    print(file_name_csv)
    
    length_input = len(input_txt_data)
    
    for i in range(length_input):
        
        hold = input_txt_data[i].split()
        
        if i == 0:
            
            jan = hold[2:]
            
        elif i == 1:
            
            feb = hold[2:]
            
        elif i == 2:
            
            mar = hold[2:]
            
        elif i == 3:
            
            apr = hold[2:]
            
        elif i == 4:
            
            may = hold[2:]
            
        elif i == 5:
            
            jun = hold[2:]
            
        elif i == 6:
            
            jul = hold[2:]
            
        elif i == 7:
            
            aug = hold[2:]
            
        elif i == 8:
            
            sep = hold[2:]
            
        elif i == 9:
            
            oct = hold[2:]
            
        elif i == 10:
            
            nov = hold[2:]
            
        else:
            
            dec = hold[2:]
            
    df_temp = pd.DataFrame({'Jan': jan, \
                            'Feb': feb, \
                            'Mar': mar, \
                            'Apr': apr, \
                            'May': may, \
                            'Jun': jun, \
                            'Jul': jul, \
                            'Aug': aug, \
                            'Sep': sep, \
                            'Oct': oct, \
                            'Nov': nov, \
                            'Dec': dec})
    
    df_temp.to_csv(file_name_csv, index=False)
        
    return    
```

```python
stationIDs = ['USC00131319', 'USW00003947', 'USW00093822']

for stationID in stationIDs:
    
    input_txt_files = ['dly_tavg_normal_'+stationID+'.txt', \
                       'dly_tmax_normal_'+stationID+'.txt', \
                       'dly_tmin_normal_'+stationID+'.txt', \
                       'dly_tavg_stddev_'+stationID+'.txt', \
                       'dly_tmax_stddev_'+stationID+'.txt', \
                       'dly_tmin_stddev_'+stationID+'.txt', \
                      ]

    for file in input_txt_files:
    
        length_file_name = len(file)
    
        end_index = length_file_name - 4
    
        substring = file[0: end_index]
    
        multiline_text_data = read_multiline_text_file(file)

        convert_to_dataframe(multiline_text_data, substring)
```

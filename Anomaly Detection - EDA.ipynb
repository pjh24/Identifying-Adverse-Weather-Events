```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Read daily climo data files: tmax & tmin  
file_climo_CedarRapids = 'Data/dly_tmax_tmin_climo_USC00131319.csv'
file_climo_KansasCity = 'Data/dly_tmax_tmin_climo_USW00003947.csv'
file_climo_Springfield = 'Data/dly_tmax_tmin_climo_USW00093822.csv'


df_climo_CedarRapids = pd.read_csv(file_climo_CedarRapids)
df_climo_KansasCity = pd.read_csv(file_climo_KansasCity)
df_climo_Springfield = pd.read_csv(file_climo_Springfield)
```

```python
#Read daily summary data files: tmax & tmin   
file_dly_CedarRapids_tmax_tmin = 'Data/dly_tmax_tmin_2023_USC00131319_cleaned.csv'
file_dly_KansasCity_tmax_tmin = 'Data/dly_tmax_tmin_2023_USW00003947_cleaned.csv'
file_dly_Springfield_tmax_tmin = 'Data/dly_tmax_tmin_2023_USW00093822_cleaned.csv'

df_dly_CedarRapids = pd.read_csv(file_dly_CedarRapids_tmax_tmin)
df_dly_KansasCity = pd.read_csv(file_dly_KansasCity_tmax_tmin)
df_dly_Springfield = pd.read_csv(file_dly_Springfield_tmax_tmin)
```

```python
print(df_dly_CedarRapids.columns)
```

```python
#Plot 30-year climo of Tmax: mean and standard deviation 
fig, axs = plt.subplots(2, 1, figsize=(8, 12))

axs[0].plot(df_climo_CedarRapids['Tmax_norm'], label='Cedar Rapids, IA')
axs[0].plot(df_climo_KansasCity['Tmax_norm'], label='Kansas City, MO')
axs[0].plot(df_climo_Springfield['Tmax_norm'], label='Springfield, IL')
axs[0].set_title('30-yr Tmax Climo: Mean ')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Degrees F')
axs[0].legend()

axs[1].plot(df_climo_CedarRapids['Tmax_std'], label='Cedar Rapids, IA')
axs[1].plot(df_climo_KansasCity['Tmax_std'], label='Kansas City, MO')
axs[1].plot(df_climo_Springfield['Tmax_std'], label='Springfield, IL')
axs[1].set_title('30-yr Tmax Climo: Standard Deviation')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Degrees F')
axs[1].legend()
```

```python
#Plot 30-year climo of Tmin: mean and standard deviation 
fig, axs = plt.subplots(2, 1, figsize=(8, 12))

axs[0].plot(df_climo_CedarRapids['Tmin_norm'], label='Cedar Rapids, IA')
axs[0].plot(df_climo_KansasCity['Tmin_norm'], label='Kansas City, MO')
axs[0].plot(df_climo_Springfield['Tmin_norm'], label='Springfield, IL')
axs[0].set_title('30-yr Tmin Climo: Mean ')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Degrees F')
axs[0].legend()

axs[1].plot(df_climo_CedarRapids['Tmin_std'], label='Cedar Rapids, IA')
axs[1].plot(df_climo_KansasCity['Tmin_std'], label='Kansas City, MO')
axs[1].plot(df_climo_Springfield['Tmin_std'], label='Springfield, IL')
axs[1].set_title('30-yr Tmin Climo: Standard Deviation')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Degrees F')
axs[1].legend()
```

```python
#Plot timeseries of Tmax for specified year 
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].plot(df_dly_CedarRapids['TMAX'], label='Cedar Rapids, IA')
axs[0].set_title('Daily Tmax for 2023 ')
#axs[0].set_xlabel('Time')
axs[0].set_ylabel('Degrees F')
axs[0].legend()

axs[1].plot(df_dly_KansasCity['TMAX'], label='Kansas City, MO')
#axs[1].set_title('30-yr Tmin Climo: Standard Deviation')
#axs[1].set_xlabel('Time')
axs[1].set_ylabel('Degrees F')
axs[1].legend()

axs[2].plot(df_dly_Springfield['TMAX'], label='Springfield, IL')
#axs[2].set_title('30-yr Tmin Climo: Standard Deviation')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Degrees F')
axs[2].legend()
```

```python
#Plot timeseries of Tmin for specified year 
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].plot(df_dly_CedarRapids['TMIN'], label='Cedar Rapids, IA')
axs[0].set_title('Daily Tmin for 2023 ')
#axs[0].set_xlabel('Time')
axs[0].set_ylabel('Degrees F')
axs[0].legend()

axs[1].plot(df_dly_KansasCity['TMIN'], label='Kansas City, MO')
#axs[1].set_title('30-yr Tmin Climo: Standard Deviation')
#axs[1].set_xlabel('Time')
axs[1].set_ylabel('Degrees F')
axs[1].legend()

axs[2].plot(df_dly_Springfield['TMIN'], label='Springfield, IL')
#axs[2].set_title('30-yr Tmin Climo: Standard Deviation')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Degrees F')
axs[2].legend()
```

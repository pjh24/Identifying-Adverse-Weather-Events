```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
stationID = 'USC00131319' #Cedar Rapids, IA
#stationID = 'USW00003947' #Kansas City, MO 
#stationID = 'USW00093822' #Springfield, IL 
```

```python
if stationID == 'USC00131319':
    
    stations_file = 'Data/3672943.csv'
    
elif stationID == 'USW00003947': 
    
    stations_file = 'Data/3674748.csv'

else:
    
    stations_file = 'Data/3674758.csv'


print(stations_file)
df_stations = pd.read_csv(stations_file)
```

```python
print(df_stations.head())
```

```python
print(df_stations.tail())
```

```python
df_station = df_stations[df_stations['STATION'] == stationID]
```

```python
print(df_station.head())
```

```python
df_station_minMax = df_station[['DATE', 'TMAX', 'TMIN']]
```

```python
print(df_station_minMax.head())
```

```python
print(df_station_minMax.dtypes)
```

```python
df_station_minMax['DATE'] = pd.to_datetime(df_station_minMax['DATE'])
```

```python
df_station_minMax['TMAX'] = df_station_minMax['TMAX'].astype(float)
df_station_minMax['TMIN'] = df_station_minMax['TMIN'].astype(float)
```

```python
print(df_station_minMax.dtypes)
```

```python
print(df_station_minMax.isna().sum())
```

```python
print(df_station_minMax[df_station_minMax.isna().any(axis=1)])
```

```python
df_station_minMax.plot(x='DATE',y='TMIN', kind='line', marker='', linestyle='-')
plt.show()
```

```python
df_station_minMax.plot(x='DATE',y='TMAX', kind='line', marker='', linestyle='-')
plt.show()
```

```python
output_year = str(df_station_minMax['DATE'].iloc[0].year)
output_filename = 'Data/dly_tmax_tmin_'+output_year+'_'+stationID+'_cleaned.csv' 
print(output_filename)
```

```python
print(df_station_minMax.dtypes)
```

```python
df_station_minMax.to_csv(output_filename, index=False)
```

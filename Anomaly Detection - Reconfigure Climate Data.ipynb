```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
stationIDs = ['USC00131319', 'USW00003947', 'USW00093822']

for stationID in stationIDs:
    
    #Read daily climo data files: tmax & tmin 
    file_tmaxNorm = 'Data/dly_tmax_normal_'+stationID+'_cleaned.csv'
    file_tmaxSTD = 'Data/dly_tmax_stddev_'+stationID+'_cleaned.csv'
    file_tminNorm = 'Data/dly_tmin_normal_'+stationID+'_cleaned.csv'
    file_tminSTD = 'Data/dly_tmin_stddev_'+stationID+'_cleaned.csv'
    
    df_tmaxNorm = pd.read_csv(file_tmaxNorm)
    df_tmaxSTD = pd.read_csv(file_tmaxSTD)
    df_tminNorm = pd.read_csv(file_tminNorm)
    df_tminSTD = pd.read_csv(file_tminSTD)
    
    #Reconfigure climo data
    df_jan_tmax_norm = df_tmaxNorm['Jan'].dropna() 
    df_jan_tmax_std = df_tmaxSTD['Jan'].dropna() 
    df_jan_tmin_norm = df_tminNorm['Jan'].dropna() 
    df_jan_tmin_std = df_tminSTD['Jan'].dropna() 

    df_feb_tmax_norm = df_tmaxNorm['Feb'].dropna()  
    df_feb_tmax_std = df_tmaxSTD['Feb'].dropna() 
    df_feb_tmin_norm = df_tminNorm['Feb'].dropna() 
    df_feb_tmin_std = df_tminSTD['Feb'].dropna() 

    df_mar_tmax_norm = df_tmaxNorm['Mar'].dropna()  
    df_mar_tmax_std = df_tmaxSTD['Mar'].dropna() 
    df_mar_tmin_norm = df_tminNorm['Mar'].dropna() 
    df_mar_tmin_std = df_tminSTD['Mar'].dropna() 

    df_apr_tmax_norm = df_tmaxNorm['Apr'].dropna()  
    df_apr_tmax_std = df_tmaxSTD['Apr'].dropna() 
    df_apr_tmin_norm = df_tminNorm['Apr'].dropna() 
    df_apr_tmin_std = df_tminSTD['Apr'].dropna() 

    df_may_tmax_norm = df_tmaxNorm['May'].dropna()  
    df_may_tmax_std = df_tmaxSTD['May'].dropna() 
    df_may_tmin_norm = df_tminNorm['May'].dropna() 
    df_may_tmin_std = df_tminSTD['May'].dropna() 

    df_jun_tmax_norm = df_tmaxNorm['Jun'].dropna()  
    df_jun_tmax_std = df_tmaxSTD['Jun'].dropna() 
    df_jun_tmin_norm = df_tminNorm['Jun'].dropna() 
    df_jun_tmin_std = df_tminSTD['Jun'].dropna() 

    df_jul_tmax_norm = df_tmaxNorm['Jul'].dropna()  
    df_jul_tmax_std = df_tmaxSTD['Jul'].dropna() 
    df_jul_tmin_norm = df_tminNorm['Jul'].dropna() 
    df_jul_tmin_std = df_tminSTD['Jul'].dropna() 

    df_aug_tmax_norm = df_tmaxNorm['Aug'].dropna()  
    df_aug_tmax_std = df_tmaxSTD['Aug'].dropna() 
    df_aug_tmin_norm = df_tminNorm['Aug'].dropna() 
    df_aug_tmin_std = df_tminSTD['Aug'].dropna() 

    df_sep_tmax_norm = df_tmaxNorm['Sep'].dropna()  
    df_sep_tmax_std = df_tmaxSTD['Sep'].dropna() 
    df_sep_tmin_norm = df_tminNorm['Sep'].dropna() 
    df_sep_tmin_std = df_tminSTD['Sep'].dropna() 

    df_oct_tmax_norm = df_tmaxNorm['Oct'].dropna()  
    df_oct_tmax_std = df_tmaxSTD['Oct'].dropna() 
    df_oct_tmin_norm = df_tminNorm['Oct'].dropna() 
    df_oct_tmin_std = df_tminSTD['Oct'].dropna() 

    df_nov_tmax_norm = df_tmaxNorm['Nov'].dropna()  
    df_nov_tmax_std = df_tmaxSTD['Nov'].dropna() 
    df_nov_tmin_norm = df_tminNorm['Nov'].dropna() 
    df_nov_tmin_std = df_tminSTD['Nov'].dropna() 

    df_dec_tmax_norm = df_tmaxNorm['Dec'].dropna()  
    df_dec_tmax_std = df_tmaxSTD['Dec'].dropna() 
    df_dec_tmin_norm = df_tminNorm['Dec'].dropna() 
    df_dec_tmin_std = df_tminSTD['Dec'].dropna() 
    
    #Concat Vertically
    df_year_tmax_norm = pd.concat([df_jan_tmax_norm, \
                                   df_feb_tmax_norm, \
                                   df_mar_tmax_norm, \
                                   df_apr_tmax_norm, \
                                   df_may_tmax_norm, \
                                   df_jun_tmax_norm, \
                                   df_jul_tmax_norm, \
                                   df_aug_tmax_norm, \
                                   df_sep_tmax_norm, \
                                   df_oct_tmax_norm, \
                                   df_nov_tmax_norm, \
                                   df_dec_tmax_norm], \
                                   ignore_index=True)

    df_year_tmax_std = pd.concat([df_jan_tmax_std, \
                                  df_feb_tmax_std, \
                                  df_mar_tmax_std, \
                                  df_apr_tmax_std, \
                                  df_may_tmax_std, \
                                  df_jun_tmax_std, \
                                  df_jul_tmax_std, \
                                  df_aug_tmax_std, \
                                  df_sep_tmax_std, \
                                  df_oct_tmax_std, \
                                  df_nov_tmax_std, \
                                  df_dec_tmax_std], \
                                  ignore_index=True)

    df_year_tmin_norm = pd.concat([df_jan_tmin_norm, \
                                   df_feb_tmin_norm, \
                                   df_mar_tmin_norm, \
                                   df_apr_tmin_norm, \
                                   df_may_tmin_norm, \
                                   df_jun_tmin_norm, \
                                   df_jul_tmin_norm, \
                                   df_aug_tmin_norm, \
                                   df_sep_tmin_norm, \
                                   df_oct_tmin_norm, \
                                   df_nov_tmin_norm, \
                                   df_dec_tmin_norm], \
                                    ignore_index=True)

    df_year_tmin_std = pd.concat([df_jan_tmin_std, \
                                  df_feb_tmin_std, \
                                  df_mar_tmin_std, \
                                  df_apr_tmin_std, \
                                  df_may_tmin_std, \
                                  df_jun_tmin_std, \
                                  df_jul_tmin_std, \
                                  df_aug_tmin_std, \
                                  df_sep_tmin_std, \
                                  df_oct_tmin_std, \
                                  df_nov_tmin_std, \
                                  df_dec_tmin_std], \
                                  ignore_index=True)
    
    #Concat horizontally
    df_year_climo = pd.concat([df_year_tmax_norm, df_year_tmax_std, df_year_tmin_norm, df_year_tmin_std], \
                          axis=1)

    df_year_climo = df_year_climo.rename(columns={0:'Tmax_norm', 1:'Tmax_std', 2:'Tmin_norm', 3:'Tmin_std'})
    
    #Output 
    outfile = 'Data/dly_tmax_tmin_climo_'+stationID+'.csv'
    print(outfile)
    df_year_climo.to_csv(outfile, index=False)
    
```
